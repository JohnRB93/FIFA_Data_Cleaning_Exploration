```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\john_\projects\FIFA_Data_Cleaning_Exploration\fifa21 raw data v2.csv', low_memory=False)
```

# Data Cleaning and Exploration

```python
pd.set_option('display.max_columns', 77)
df.head()
```

```python
# Many of the player and photo urls are old and don't even work anymore, so remove those two columns
df.drop(columns=['photoUrl', 'playerUrl'], inplace=True)
df.head()
```

```python
# Remove the '\n' characters in the 'Club' column
df['Club'] = df['Club'].str.replace('\n', '')
df.head()
```

```python
# Rename the '↓OVA' column to 'OVA
df.rename(columns={'↓OVA': 'OVA'}, inplace=True)
df.head()
```

```python
# Split the 'Contract' column into two separate columns, 'Contract Year Start' and 'Contract Year End'
df[['Contract Year Start', 'Contract Year End']] = df['Contract'].str.split(pat=' ~ ', n=1, expand=True)
df = df.drop(columns='Contract')
df.head()
```

```python
# Change the position of the newly added 'Contract Year Start' and 'Contract Year End' columns
columns = df.columns.tolist()
columns.insert(columns.index('Joined'), columns.pop(columns.index('Contract Year Start')))
columns.insert(columns.index('Joined'), columns.pop(columns.index('Contract Year End')))
df = df[columns]
df.head()
```

```python
# Remove the '★' characters from the 'IR', 'W/F', and 'SM' columns
df['IR'] = df['IR'].str.replace(pat=' ★', repl='')
df['W/F'] = df['W/F'].str.replace(pat=' ★', repl='')
df['SM'] = df['SM'].str.replace(pat='★', repl='')
df.head()
```

```python
# The 'Value', 'Wage', and 'Release Clause' columns are filled with string values, convert them into numerical values
# Remove the M and K units at the end of the values and replace the value with the correct numbers
# ex. 302.5M -> 302500000     302.5K -> 302500

def money_to_nums(val):
    if 'M' in val:
        return float(val.replace('€', '').replace('M', '')) * 1_000_000
    elif 'K' in val:
        return float(val.replace('€', '').replace('K', '')) * 1_000
    else:
        return None

df['Value'] = df['Value'].apply(money_to_nums)
df['Wage'] = df['Wage'].apply(money_to_nums)
df['Release Clause'] = df['Release Clause'].apply(money_to_nums)
df.head()
```

```python
# Find all the different units of measurement for the 'Height' column and use the same unit of measurement (cm) for each column
# Find all the different units of measurement for the 'Weight' column and use the same unit of measurement (kg) for each column

def inches_to_cm(val):
    return 2.54 * val

def feet_to_cm(val):
    return inches_to_cm(val * 12)

def convert_to_cm(val):
    if '\'' in val:
        feet_in_cm = feet_to_cm(int(val[0]))
        inches_in_cm = inches_to_cm(int(val[2: -1]))
        return feet_in_cm + inches_in_cm
    else:
        return float(val.replace('cm', ''))

def convert_to_kg(val):
    if 'lbs' in val:
        float_val = float(val.replace('lbs', ''))
        return float_val * 0.453592
    else:
        float_val = float(val.replace('kg', ''))
        return float_val

df['Height'] = df['Height'].apply(convert_to_cm).astype(np.float32)
df['Weight'] = df['Weight'].apply(convert_to_kg).astype(np.float32)
df.head()
```

```python
def extract_year(date):
    if ',' in date:
        return date.split(', ')[1].replace(' On Loan', '')
    else:
        return date

df['Contract Year Start'] = df['Contract Year Start'].apply(extract_year)
df.head()
```

```python
# Check if there are any more null values to clean

pd.set_option('display.max_rows', None)
df.isna().sum()
```

```python
# Clean nulls in the 'Hits' column and convert to numeric value

def to_num(val):
    if 'K' in val:
        int_val = int(val.replace('K', '').replace('.', '')) * 1_000
        return int_val
    else:
        return int(val)

df['Hits'] = df['Hits'].fillna('0').apply(to_num)
df.head()
```

```python
# Break the 'Positions' column into separate columns
df[['Position 1', 'Position 2', 'Position 3']] = df['Positions'].str.split(', ', expand=True)
df.drop(columns=['Positions'], inplace=True)
df.head()
```

```python
# Move the new position columns before the 'Best Postition' column
columns = df.columns.tolist()
columns.insert(columns.index('Best Position'), columns.pop(columns.index('Position 1')))
columns.insert(columns.index('Best Position'), columns.pop(columns.index('Position 2')))
columns.insert(columns.index('Best Position'), columns.pop(columns.index('Position 3')))
df = df[columns]
df.head()
```
